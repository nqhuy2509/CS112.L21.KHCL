# Abtraction: 

-Nhập 2 số nguyên n,k
-tìm F(k) theo mô đun 10^9 +7:
  + khi k=0 :n
  + khi k=1 :n
  + khi k=2 :3n
  + khi k>2 : 3f(k-1) -f(k-2)

# Partern recognition:
Dạng bài toán Fibonaci:

    F(n) = F(n-1) + F(n-2)

#Algorithm designed:
- Input n , k
- f1=0
- f2=n
- fk=2*n
- for _ in range(1,2k):
  + f1=f2
  + f2=fk
  + fk=3*f2-f1

print fk % 10**9+7
