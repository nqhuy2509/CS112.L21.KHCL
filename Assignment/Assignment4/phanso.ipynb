1. **Abtraction**:
- Biến đổi phân số a/b thành phân số c/d sau mỗi lần tăng a, b lên 1
- Tối giản phân số sau những lần biến đổi
- Tìm điều kiện dừng nếu không có phân số c/d phù hợp


2. **Pattern Recognition**:
- Tìm phân số tối giản
- Tìm điều kiện dừng bài toán


Algorithm Design:

- GCD (a,b):

    if b ==0: return a

    return GCD( b, a % b )
- main:
  loop(a <> c or b<> d):
   + a,b +=1
   + a,b /= GCD(a,b)
   + if a== b-1 and a > c: count=0, break
   + else count+=1

Code :


 


```python
# Code by Nguyen quan huy

a=int(input())
b=int(input())
c=int(input())
d=int(input())

def uocso(a,b):
    if (b == 0):
        return a;
    return uocso(b, a % b);

count=0
while(a!=c or b!=d):
    a += 1
    b += 1
    x=uocso(a,b)
    a=a/x
    b=b/x
    if a==b-1:
        if a>c:
            count=0
            break
    count+=1

print(count)
```
